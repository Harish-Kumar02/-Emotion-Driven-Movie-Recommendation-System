```python
import h5py
from sklearn.decomposition import TruncatedSVD
import joblib
```

```python
# Load the cosine similarity matrix
with h5py.File('cosine_sim_matrix.h5', 'r') as hdf:
    cosine_sim = hdf['cosine_sim_matrix'][:]
```

```python
# Apply SVD for dimensionality reduction (Collaborative Filtering)
svd = TruncatedSVD(n_components=50)
svd_matrix = svd.fit_transform(cosine_sim)

# Save the SVD matrix
with h5py.File('svd_matrix.h5', 'w') as hdf:
    hdf.create_dataset('svd_matrix', data=svd_matrix)

# Save the SVD model using joblib
joblib.dump(svd, 'svd_model.pkl')

print("SVD computation completed and model saved.")
```
