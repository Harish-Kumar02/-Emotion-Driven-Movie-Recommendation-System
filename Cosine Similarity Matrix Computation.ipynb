```python
import h5py
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import vstack
```

```python
# Load the TF-IDF matrix
with h5py.File('tfidf_matrix.h5', 'r') as hdf:
    tfidf_matrix = hdf['tfidf_matrix'][:]
```

```python
# Compute cosine similarity in chunks
chunk_size = 100  # Adjust based on your available memory
cosine_sim_list = []

for i in range(0, tfidf_matrix.shape[0], chunk_size):
    cosine_chunk = cosine_similarity(tfidf_matrix[i:i+chunk_size], tfidf_matrix)
    cosine_sim_list.append(cosine_chunk)

# Stack the cosine similarity chunks together
cosine_sim = np.vstack(cosine_sim_list)

# Save the cosine similarity matrix
with h5py.File('cosine_sim_matrix.h5', 'w') as hdf:
    hdf.create_dataset('cosine_sim_matrix', data=cosine_sim)

print("Cosine similarity matrix saved.")
```
